```python
import matplotlib.pyplot as plt
import numpy as np

# config
x_names   = ['16', '32', '128']
y1_values = [0, 0, 0.9897]
y2_values = [0.6762, 0.8829, 0.9904]
y3_values = [0.8562, 0.9642, 0.9982]
y4_values = [0, 0, 0.9981]
y5_values = [0.8819, 0.9978, 0.9985]
y6_values = [0, 0.8975, 0.9982]
y7_values = [0.8346, 0.9972, 0.9983]
y8_values = [0.8237, 0.9974, 0.998]

size_of_group = 8

# plot
width = 0.85*1/size_of_group
x = np.arange(len(x_names))
y = np.linspace(0,1,5)
fig, ax = plt.subplots()
fig.set_size_inches(17, 5)

rects1 = ax.bar(x - 3.5*width, y1_values, width, label='NN')
rects2 = ax.bar(x - 2.5*width, y2_values, width, label='Smooth NN')
rects3 = ax.bar(x - 1.5*width, y3_values, width, label='Minimax')
rects4 = ax.bar(x - 0.5*width, y4_values, width, label='SINDy')
rects5 = ax.bar(x + 0.5*width, y5_values, width, label='Lattice')
rects6 = ax.bar(x + 1.5*width, y6_values, width, label='GP')
rects7 = ax.bar(x + 2.5*width, y7_values, width, label='PIGP1')
rects8 = ax.bar(x + 3.5*width, y8_values, width, label='PIGP2')

ax.set_xlabel('Training dataset volumn', fontsize=16)
ax.set_ylabel(r'$R^2$', fontsize=16)
ax.set_xticks(x, x_names, fontsize=16)
ax.set_yticks(y, y, fontsize=16)
ax.legend(loc='best', fontsize=12)

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.bar_label(rects3, padding=3)
ax.bar_label(rects4, padding=3)
ax.bar_label(rects5, padding=3)
ax.bar_label(rects6, padding=3)
ax.bar_label(rects7, padding=3)
ax.bar_label(rects8, padding=3)

fig.tight_layout()

plt.savefig('r2compare.pdf') 
plt.show()
```
